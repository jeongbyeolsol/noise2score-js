```python
import argparse
import shutil
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_group(df, x, columns, title, ylabel, save_path):
    plt.figure(figsize=(10, 5))

    for col in columns:
        if col in df.columns:
            plt.plot(df[x], df[col], label=col)

    plt.xlabel(x)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig(save_path, dpi=200)
    plt.close()
```

```python
name = 'poisson_004'

parent_path =  Path("./checkpoints/ardae/") / Path(name)
csv_path = parent_path / Path("metrics.csv")
out_dir = Path("./plots/") / Path(name)
out_dir.mkdir(parents=True, exist_ok=True)

df = pd.read_csv(csv_path)

# epoch 기준 정렬
df = df.sort_values("epoch")

config_path = parent_path / Path('config.json')
metrics_path = parent_path / Path('metrics.csv')
shutil.copy(config_path, out_dir)
shutil.copy(metrics_path, out_dir)
```

```python
# 1. loss
plot_group(
        df,
        x="epoch",
        columns=["train_loss", "val_loss", "best_val_loss"],
        title="Loss",
        ylabel="loss",
        save_path=out_dir / "loss.png",
    )
```

```python
# 2. Score MSE / NMSE
plot_group(
        df,
        x="epoch",
        columns=[
            "train_score_mse",
            "val_score_mse",
            "train_score_nmse",
            "val_score_nmse",
        ],
        title="Score MSE / NMSE",
        ylabel="error",
        save_path=out_dir / "score_error.png",
    )
```

```python
# 3. Cosine similarity
plot_group(
        df,
        x="epoch",
        columns=["train_score_cos", "val_score_cos"],
        title="Score Cosine Similarity",
        ylabel="cosine similarity",
        save_path=out_dir / "score_cos.png",
    )
```

```python
# 4. Correlation
plot_group(
        df,
        x="epoch",
        columns=["train_score_corr", "val_score_corr"],
        title="Score Correlation",
        ylabel="correlation",
        save_path=out_dir / "score_corr.png",
    )
```

```python
#  5. Energy
plot_group(
        df,
        x="epoch",
        columns=[
            "train_target_energy",
            "train_pred_energy",
            "val_target_energy",
            "val_pred_energy",
        ],
        title="Target / Pred Energy",
        ylabel="energy",
        save_path=out_dir / "energy.png",
    )
```
